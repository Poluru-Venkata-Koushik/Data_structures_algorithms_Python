`Program to read an Image`

```python
# importing all necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_in=cv2.imread('/content/image_in.jpg')
img_rgb = cv2.cvtColor(img_in, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('Off')
```

**Printing Red, Green and Blue Channel Matrices**

```python
r,g,b=cv2.split(img_rgb)
print("R Channel :\n",r)
print("G Channel :\n",g)
print("B Channel :\n",b)
```

***Extracting Channel Images***

```python
plt.imshow(r)
plt.title("R Channel Image")
plt.axis('Off')
```

```python
plt.imshow(g)
plt.title("G Channel Image")
plt.axis('Off')
```

```python
plt.imshow(b)
plt.title("B Channel Image")
plt.axis('Off')
```

***Merging channels to verify if the re form the original image***

```python
plt.imshow(cv2.merge([r,g,b]))
plt.title("Merging all Channels")
plt.axis('Off')
```
